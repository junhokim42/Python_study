#### 관련 문법 : SCIPY.SPATIAL.DISTANCE.CDIST

- 두 개의 행렬을 바탕으로 거리 행렬을 반환하는 함수
- scipy.spatial.distance.cdist(XA, XB, metric='euclidean', *, out=None, **kwargs)
    - XA : 거리 행렬 계산 대상인 행렬(ndarray or DF)로 함수의 출력에 행에 해당
    - XB : 거리 행렬 계산 대상인 행렬(ndarray or DF)로 함수의 출력의 열에 해당
    - metric : 거리척도계산 함수('cityblock','correlation','cosine','euclidean','jaccard','matching' 등)
    
    



#### 하버사인 (haversine) 패키지
- 파이썬 라이브러리 중에 위경도 값 데이터를 이용해서 두 지점 간 거리를 구할 때 유용하게 쓸 수 있는 패키지

- 패키지 필요성
    - 평면에서 두 지점간의 거리 계산
        - 피타고라스 정리 사용
    
    
    - 지구는 평면이 아니고 둥글고 최단거리는 직선이 아닌 곡선으로 생성됨
        - 피타고라스 정리 사용 불가
        - 곡선의 거리 구하는 공식 : 하버 사인(Haversine) 공식
 

#### 이웃을 찾는데 활용되는 함수 : ndarray.argsort(axis=0/1) : axis=0 이면 열별 위치를 1이면 행별 위치를 반환
    - axis=1 default

- 작은 값부터 순서대로 데이터의 위치를 반환하는 함수
