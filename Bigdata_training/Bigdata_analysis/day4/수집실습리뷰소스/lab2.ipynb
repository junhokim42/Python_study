```python
# CASE1
import requests
from bs4 import BeautifulSoup

req = requests.get('http://unico2013.dothome.co.kr/crawling/exercise_bs.html')
html = req.content
#html = html.decode('utf-8')
#print(html)
#print("-"*60)
bs = BeautifulSoup(html, 'html.parser')

print("[<h1> 태그의 콘텐츠]", bs.h1.string)
print("[텍스트 형식으로 내용을 가지고 있는 <a> 태그의 콘텐츠와 href 속성값]")
for i in bs.select('a'):
    if i.get_text(strip=True):
        print(i.text, ":",i["href"])
print("[<img>태그의 src 속성값]", bs.select('img')[0]['src'])
print("[첫 번째 <h2>태그의 콘텐츠]", bs.h2.string)
print("[<ul> 태그의 자식 태그들 중 style 속성의 값이 green으로 끝나는 태그의 콘텐츠]", bs.find('li', attrs={"style":"color:green"}).string)
print("[두 번째 <h2>태그의 콘텐츠]", bs.select("h2:nth-of-type(2)")[0].string)
print("[<ol>태그의 모든 자식 태그들의 콘텐츠]", bs.ol.text)
print("[<table> 태그의 모든 자손 태그들의 콘텐츠]", bs.table.text)
print("[name이라는 클래스 속성을 갖는 <tr> 태그의 컨텐츠]", bs.find('tr', attrs={"class":"name"}).text)
print("[target이라는 아이디 속성을 갖는 <td> 태그의 콘텐츠]", bs.find('td', attrs={"id":"target"}).text)
```

```python
# CASE2

import requests
from bs4 import BeautifulSoup

req = requests.get('http://unico2013.dothome.co.kr/crawling/exercise_bs.html')
html = req.content
html = html.decode('utf-8')
bs = BeautifulSoup(html, 'html.parser')

#1
print("[<h1>태그의 컨텐츠]",bs.find('h1').text)
print()

#2
contents = bs.find_all('a',text=True)
print("[텍스트 형식으로 내용을 가지고 있는 <a> 태그의 콘텐츠와 href 속성값]")
for content in contents :
    print(content.text+" : "+ content["href"])
print()
    
#3
content = bs.find('img')
print("[<img> 태그의 src 속성값]",content['src'])
print()

#4
content = bs.find('h2')
print("[첫 번째 <h2> 태그의 콘텐츠]",content.text)
print()

#5 
import re
#print(bs.ul.contents)
content = bs.find_all('li',attrs={'style':re.compile('green$')})
print("[<ul> 태그의 자식 태그들 중 style 속성의 값이 green으로 끝나는 태그의 콘텐츠]",content[0].text)
print()

#6
content = bs.select('h2:nth-of-type(2)')
print("[두 번째 <h2> 태그의 콘텐츠]",content[0].text)
print()

#7
contents = bs.ol
print("[<ol> 태그의 모든 자식 태그들의 콘텐츠]",contents.text)
print()

#8 
contents = bs.table
print("[<table> 태그의 모든 자손 태그들의 콘텐츠]",contents.text)
print()

#9
content = bs.find('tr',class_="name")
print("[name이라는 클래스 속성을 갖는 <tr> 태그의 콘텐츠]",content.text)
print()

#10
content = bs.find('td',id="target")
print("[target이라는 아이디 속성을 갖는 <td> 태그의 콘텐츠]",content.text)
print()
```
