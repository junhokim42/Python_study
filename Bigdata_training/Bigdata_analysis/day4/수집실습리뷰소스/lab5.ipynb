```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys 
import pandas as pd
import re, time

driver = webdriver.Chrome(service=Service("c:/Temp/chromedriver"))
driver.implicitly_wait(3)

url = "http://gs25.gsretail.com/gscvs/ko/products/event-goods"
driver.get(url)
time.sleep(3)

# 2+1 행사 코너로 이동
btn = driver.find_element(By.ID, "TWO_TO_ONE")
btn.click()

# 상품명, 가격 스크래핑
goodsname = []
goodsprice = []

for page in range(10):
    print(f"{page+1} 페이지 스크래핑...")
    time.sleep(2)
    tit = driver.find_elements(By.CSS_SELECTOR, "div:nth-child(5) p.tit")
    price = driver.find_elements(By.CSS_SELECTOR, "div:nth-child(5) p.price")    

    for i in range(len(tit)):
        goodsname.append(tit[i].text)
        temp_price = re.sub("[^0-9]", "", price[i].text)
        goodsprice.append(temp_price)
    
    btn_next = driver.find_element(By.LINK_TEXT, "다음 페이지로 이동")
    btn_next.click()
print(f"<{len(goodsname)}개 아이템 스크래핑 완료>")

# csv 파일로 저장
goods_dict = {"goodsname":goodsname, "goodsprice":goodsprice}
df = pd.DataFrame(goods_dict)
display(df)
df.to_csv("output/gs25_twotoone.csv", index=False)
```
