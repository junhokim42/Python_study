```python
seoul_commute_in = seoul_parking_4[(seoul_parking_4['통근통학지']=='서울특별시')]
seoul_commute_in
```

```python
seoul_parking_5 = pd.read_csv('./unite/data/기타지역에서_서울통근통학지별.csv')
seoul_ex = seoul_parking_5.sum(axis=0)
seoul_ex = seoul_ex.to_frame()
seoul_ex = seoul_ex.T
seoul_ex.columns = ['현거주지', '통근통학지', '통근통학인구(12세 이상)']
seoul_ex.loc[0, '현거주지'] = '기타'
seoul_ex.loc[0, '통근통학지'] = '서울특별시'
seoul_ex
seoul_commute_inT = pd.concat([seoul_commute_in, seoul_ex])
seoul_commute_inT
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(25,10))
plt.title('서울시 통근통학 인구수', size=20)

index = np.arange(seoul_commute_out.shape[0])
w = 0.3

plt.bar(index-w, seoul_commute_out['통근통학인구(12세 이상)'], width=w, color = ['b'], alpha = 0.4, label='서울에서 나가는 인구수')
plt.bar(index, seoul_commute_inT['통근통학인구(12세 이상)'], width=w, color = ['g'], alpha = 0.4, label='서울에 들어오는 인구수')

plt.legend()
plt.xticks(index, seoul_commute_out['통근통학지'])
plt.show()
```

```python
# from google.colab import drive
# drive.mount('/content/drive')

import pandas as pd
import numpy as np
seoul_parking_7 = pd.read_csv('./unite/data/서울시구별들어오는통근통학인구.csv')
seoul_parking_7.columns = [['구', '유입인구수']]
seoul_parking_8 = pd.read_csv('./unite/data/서울시구별나가는통근통학인구.csv')
seoul_parking_8.columns = [['구', '유출인구수']]
seoul_Gu_inout = pd.merge(seoul_parking_7, seoul_parking_8)

def x(a,b):
    return a - b

seoul_Gu_inout['순유입인구수']  = seoul_Gu_inout.apply(lambda f: x(f['유입인구수'],f['유출인구수']), axis=1)
seoul_Gu_inout.head()
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt
import numpy as np

plt.rc("font", family = "Malgun Gothic")
plt.rc("axes", unicode_minus = False)

plt.figure(figsize=(25,10))
plt.title('서울시 구별 통근통학 순유입인구수', size=20)

index = np.arange(seoul_Gu_inout.shape[0])
w = 0.5
k = seoul_Gu_inout['순유입인구수'].values.tolist()
klist = sum(k, [])
k2 = seoul_Gu_inout['구'].values.tolist()
klist2 = sum(k2, [])
plt.bar(index, klist, align='center', alpha=0.5)

plt.rcParams['axes.unicode_minus'] = False
plt.xticks(index, klist2)
plt.show()
```
