### [2] 파생변수의 생성
#### 1) 표현형태 변환하기

```python
#1. 모듈 및 데이터 탑재
import pandas as pd
import numpy as np
df = pd.read_csv('Ashopping.csv',sep=',', encoding='CP949')

#2. 표현형태 변환하기
df['남녀']=np.where(df.성별 == 0, '남자', '여자')

#3. 데이터 확인하기
df[['고객ID', '성별', '남녀']]
```

#### 2) 척도 변환하기

```python
#1. 모듈 및 데이터 탑재
import pandas as pd
import numpy as np
df = pd.read_csv('Ashopping.csv',sep=',', encoding='CP949')

#2. 척도 변환 파생변수 생성
df['New고객등급'] = np.where(df.총_매출액 >=5000000, '상', np.where(df.총_매출액>=3000000, '중', '하'))

#3. 데이터 확인하기
df[['고객ID','총_매출액','New고객등급']]
```

#### 3) 모델링을 활용해 파생변수 생성하기

```python
#1. 모듈 및 데이터 탑재
import pandas as pd
df = pd.read_csv('Ashopping.csv',sep=',', encoding='CP949')

#2. 척도 변환 파생변수 생성
df['New_1회_평균매출액']=df['총_매출액']/df['방문빈도'] 

#3. 데이터 확인하기
df[['고객ID', '총_매출액', '방문빈도', 'New_1회_평균매출액']]
```

#### 4) 거래이력 요약을 통한 파생변수 생성하기

```python
#1. 모듈 및 데이터 탑재
import pandas as pd
df = pd.read_csv('Ashopping.csv',sep=',', encoding='CP949')

#2. 비이탈고객 레코드 추출
churn_customer = df[df.이탈여부==1] 
Non_churn_customer = df[df.이탈여부==0]

#3. 매출합계 출력하기
print(sum(churn_customer.총_매출액))
print(sum(Non_churn_customer.총_매출액))
```

## 4.3 유용한 통계 그래프
### [1] 히스토그램

```python
#1. 모듈 및 데이터 탑재
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt 
df = pd.read_csv('Ashopping.csv',sep=',', encoding='CP949')

#2. 한글깨짐현상 방지 
matplotlib.rcParams['font.family'] = 'Malgun Gothic'   # '맑은 고딕'으로 설정 
matplotlib.rcParams['axes.unicode_minus'] = False

#3. 히스토그램 작성하기
%matplotlib inline
plt.hist(df['서비스_만족도'], alpha=0.4, bins=7, rwidth=1, color ='red', label='서비스만족도')

#4. 각종 옵션설정
plt.legend()
plt.grid()
plt.xlabel('서비스 만족도')
plt.ylabel('빈도')
plt.xticks(fontsize = 14) #글씨 크기설정
plt.yticks(fontsize = 14) #글씨 크기설정
plt.show()
```

### [3] 파이차트

```python
#1. 모듈 및 데이터 탑재
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
df = pd.read_csv('Ashopping.csv',sep=',', encoding='CP949')

#2. 한글깨짐현상 방지 
matplotlib.rcParams['font.family'] = 'Malgun Gothic'   # '맑은 고딕'으로 설정 
matplotlib.rcParams['axes.unicode_minus'] = False

#3. 구매유형별 고객 그룹화하기
groupby_구매유형 = df.groupby('구매유형')
groupby_구매유형.describe()

#4. 파이차트 속성 지정
labels = ['1','2','3','4'] #라벨 지정
sizes = [43, 317, 144, 496] # 크기 지정
colors = ['yellowgreen','gold','lightskyblue','lightcoral'] #색 지정
explode = (0, 0.1, 0, 0)

#5. 파이차트 작성하기
%matplotlib inline
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct='%1.1f%%',shadow = True, startangle = 90)
plt.axis('equal')
plt.title('구매유형에 따른 고객 분포')
plt.show()
```
